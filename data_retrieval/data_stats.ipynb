```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('songs.csv')

# Assuming the column with popularity is named 'popularity'
popularity_counts = df['popularity'].value_counts().sort_index()

# Cut songs with popularity 0
#popularity_counts = popularity_counts[50:]


# Plot the distribution
plt.figure(figsize=(10, 6))
popularity_counts.plot(kind='bar')
plt.title('Distribution of Song Popularity')
plt.xlabel('Popularity')
plt.ylabel('Number of Songs')
plt.show()
```
