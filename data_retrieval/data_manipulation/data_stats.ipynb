```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv('../datasets/random_songs_recommendations.csv')

# Assuming the column with popularity is named 'popularity'
popularity_counts = df['popularity'].value_counts().sort_index()

# print number of songs
print('Number of songs:', "{:,}".format(len(df)))

# drop duplicates
df = df.drop_duplicates(subset='id')

# Cut songs with popularity 0
popularity_counts = popularity_counts[80:]

# Plot the distribution
plt.figure(figsize=(10, 6))
popularity_counts.plot(kind='bar')
plt.title('Distribution of Song Popularity')
plt.xlabel('Popularity')
plt.ylabel('Number of Songs')

# Calculate the range for x-axis ticks
# x_ticks = np.linspace(start=min(popularity_counts.index), stop=max(popularity_counts.index), num=10, dtype=int)
# plt.xticks(ticks=x_ticks, labels=x_ticks)

plt.show()
```
